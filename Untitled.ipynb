```python
import librosa
import numpy as np
import pywt
import matplotlib.pyplot as plt
import soundfile as sf
import scipy.io.wavfile as wavfile
import librosa.display
from scipy.signal import butter, filtfilt, spectrogram
```

```python

def bandpass_filter(signal, fs, lowcut, highcut, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    filtered = filtfilt(b, a, signal)
    return filtered


def plot_spectrogram(signal, fs, title="Spectrogram"):
    plt.figure(figsize=(12, 4))
    D = librosa.amplitude_to_db(np.abs(librosa.stft(signal)), ref=np.max)
    librosa.display.specshow(D, sr=fs, x_axis='time', y_axis='hz', cmap='magma')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.tight_layout()
    plt.show()


def load_audio_librosa(filepath, sr=None):
    signal, sample_rate = librosa.load(filepath, sr=sr, mono=True)
    return signal, sample_rate


def wavelet_denoise(signal, wavelet='coif3', level=4):
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    uthresh = sigma * np.sqrt(2 * np.log(len(signal)))
    coeffs_thresh = [pywt.threshold(c, value=uthresh, mode='soft') for c in coeffs]
    return pywt.waverec(coeffs_thresh, wavelet)


def plot_signals(original, denoised, fs):
    time = np.linspace(0, len(original)/fs, num=len(original))
    plt.figure(figsize=(14, 5))
    plt.plot(time, original, label="Original", alpha=0.5)
    plt.plot(time, denoised, label="Denoised", alpha=0.7)
    plt.legend()
    plt.title("Сравнение исходного и очищенного сигнала")
    plt.xlabel("Время (сек)")
    plt.ylabel("Амплитуда")
    plt.tight_layout()
    plt.show()


def save_to_wav(filename, signal, fs):
    # Ограничим амплитуду перед сохранением
    signal = signal / np.max(np.abs(signal))
    sf.write(filename, signal, fs)
    print(f"Файл сохранён: {filename}")


# === ЗАПУСК ===
if __name__ == "__main__":
    input_file = "denoised_2.wav"
    signal, fs = load_audio_librosa(input_file)
    

    
    # Строим спектрограмму
    frequencies, times, Sxx = spectrogram(signal, fs)
    
    # Находим диапазон частот с максимальной амплитудой
    max_amplitude_freq_range = np.argmax(np.max(Sxx, axis=1))  # Частота с наибольшей амплитудой
    max_amplitude = np.max(Sxx)
    
    # Получаем диапазон частот для максимальной амплитуды
    # Например, будем смотреть на частоты в пределах 10% от максимальной амплитуды
    threshold = 0.007 * max_amplitude
    freq_range = frequencies[np.max(Sxx, axis=1) > threshold]

    denoised = wavelet_denoise(signal)

    
    # Обрезаем до совпадения длины
    min_len = min(len(signal), len(denoised))
    signal = signal[:min_len]
    denoised = denoised[:min_len]
    
    filtered = bandpass_filter(denoised, fs, lowcut=freq_range[0], highcut=freq_range[-1])

    # Визуализация временных сигналов
   # plot_signals(signal, denoised, fs)

    # Визуализация спектрограмм
    plot_spectrogram(filtered, fs, title="Исходная спектрограмма")
   # plot_spectrogram(denoised, fs, title="Спектрограмма после вейвлет-фильтрации")

    # Сохраняем результат
    # save_to_wav("denoised_output.wav", denoised, fs)
    # save_to_wav("filtered_output.wav", filtered, fs)



    
    
```

### Кластеризация-метод

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Загрузка
y, sr = librosa.load("morse-decoder/morse_dataset/morse_dataset/21.opus", sr=None)
y = y / np.max(np.abs(y))  # нормализация

# Извлечение MFCC (можно 13 или 20 коэффициентов)
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
mfcc = mfcc.T  # (временные шаги, признаки)
```

```python
# Масштабируем признаки
scaler = StandardScaler()
mfcc_scaled = scaler.fit_transform(mfcc)

# Кластеризация (k=2 предполагает "шум" и "сигнал")
kmeans = KMeans(n_clusters=2, random_state=0).fit(mfcc_scaled)
labels = kmeans.labels_
```

```python
# Временная ось
times = librosa.frames_to_time(np.arange(len(labels)), sr=sr)

plt.figure(figsize=(14, 4))
plt.plot(times, labels, drawstyle='steps-mid')
plt.title("Кластеры MFCC (0 — шум, 1 — выбросы)")
plt.xlabel("Время (сек)")
plt.ylabel("Кластер")
plt.grid()
plt.show()
```

```python
# Средняя энергия каждого кластера
energy = np.sum(librosa.feature.rms(y=y), axis=0)
energy = energy[:len(labels)]  # синхронизация по длине

cluster_energy = [np.mean(energy[labels == i]) for i in range(2)]
signal_cluster = np.argmax(cluster_energy)
```

```python
from itertools import groupby
from operator import itemgetter

# Выбросы
signal_frames = np.where(labels == signal_cluster)[0]

# Группировка подряд идущих
groups = []
for k, g in groupby(enumerate(signal_frames), lambda ix: ix[0] - ix[1]):
    group = list(map(itemgetter(1), g))
    start = librosa.frames_to_time(group[0], sr=sr)
    end = librosa.frames_to_time(group[-1], sr=sr)
    groups.append((start, end))

# Вывод
for i, (start, end) in enumerate(groups):
    print(f"Выброс {i+1}: {start:.2f}s – {end:.2f}s")
```

```python
# Создаем копию исходного сигнала
filtered_2_signal = np.zeros_like(y)

for start, end in groups:
    start_sample = int(start * sr)
    end_sample = int(end * sr)
    filtered_2_signal[start_sample:end_sample] = y[start_sample:end_sample]

# Сохраняем версию с "тишиной на месте шума"
sf.write("filtered_2_signal.wav", filtered_2_signal, sr)
```

```python
import librosa
import numpy as np
import soundfile as sf

# 1. Загрузка аудио
audio_path = 'morse-decoder/morse_dataset/morse_dataset/10010.opus'
y, sr = librosa.load(audio_path, sr=None)  # y — аудиосигнал, sr — частота дискретизации

# 2. Применение БПФ
fft_data = np.fft.fft(y)
frequencies = np.fft.fftfreq(len(fft_data), d=1/sr)

# 3. Маска частот — оставим только нужный диапазон (например, 300–3000 Гц)
mask = (np.abs(frequencies) > 250) & (np.abs(frequencies) < 700)
filtered_fft = fft_data * mask

# 4. Обратное БПФ (чтобы вернуться к аудиосигналу)
filtered_signal = np.fft.ifft(filtered_fft)

# Только вещественная часть
filtered_signal = np.real(filtered_signal)

# 5. Сохранение результата
sf.write('fft_audio.wav', filtered_signal, sr)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import soundfile as sf  # для записи .wav
from scipy.ndimage import gaussian_filter1d

# === 1. Загрузка аудиофайла ===
filename = 'denoised_2.wav'  # Замените на свой путь morse-decoder/morse_dataset/morse_dataset/40.opus    
audio, sr = librosa.load(filename, sr=None, mono=True)          # denoised_2.wav

# === 2. Расчёт энвелопа через скользящее RMS ===
def compute_envelope(signal, frame_size):
    envelope = np.sqrt(np.convolve(signal ** 2, np.ones(frame_size) / frame_size, mode='same'))
    return envelope

frame_duration = 0.05  # окно 50 мс
frame_size = int(sr * frame_duration)
envelope = compute_envelope(audio, frame_size)

sigma = 300  # Чем больше, тем сильнее сглаживание
env_gaussian = gaussian_filter1d(envelope, sigma=sigma)

# === 3. Выделение сигнала по порогу ===-----------------------------------------------
# signal — ваш одномерный numpy-массив или список
diff = np.diff(env_gaussian[4400:40000])  # Первая разность
threshold = np.mean(diff) + 1 * np.std(diff)  # Порог: можно варьировать

# Индексы, где есть скачок выше порога
impulse_indices = 0
impulse_indices = np.where(diff > threshold)[0][0]

if impulse_indices != None:
    threshold = np.max(env_gaussian[4400:12000])
else:
    threshold = np.max(env_gaussian[4400:impulse_indices-1]) * 1.1
    
signal_mask = env_gaussian > threshold
cleaned_signal = audio * signal_mask

# === 4. Визуализация ===
t = np.linspace(0, len(audio) / sr, len(audio))

plt.figure(figsize=(15, 8))

plt.subplot(3, 1, 1)
plt.plot(t, audio, label='Original Signal')
plt.title('Оригинальный сигнал')

plt.subplot(3, 1, 2)
plt.plot(t, envelope, label='Envelope', color='orange')
plt.plot(t, env_gaussian[:176128], 'b', label=f"Гауссов фильтр (σ={sigma})")
plt.axhline(threshold, color='red', linestyle='--', label='Threshold')
plt.title('Энвелоп и порог')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, cleaned_signal, label='Detected Signal', color='green')
plt.title('Выделенный сигнал')

plt.tight_layout()
plt.show()

# === 5. Сохранение результата ===
output_file = 'cleaned_signal.wav'
sf.write(output_file, cleaned_signal, sr)
print(f"✅ Сигнал сохранён в файл: {output_file}")

```

```python
# вычитание шумов

import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import soundfile as sf

# === Загрузка аудиофайла ===
y, sr = librosa.load("morse-decoder/morse_dataset/morse_dataset/21.opus")

# === Параметры ===
noise_duration = 0.2  # первые 0.3 сек считаются шумом
n_fft = 2048
hop_length = 512

# === Преобразование STFT ===
S_full = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
S_mag, S_phase = np.abs(S_full), np.angle(S_full)

# === Оценка спектра шума ===
noise_frames = int(noise_duration * sr / hop_length)
noise_profile = np.mean(S_mag[:, :noise_frames], axis=1, keepdims=True)

# === Спектральное вычитание ===
S_denoised_mag = np.maximum(S_mag - noise_profile, 0)
S_denoised = S_denoised_mag * np.exp(1j * S_phase)

# === Обратное преобразование ===
y_denoised = librosa.istft(S_denoised, hop_length=hop_length)

# === Сохранение аудио ===
#sf.write("original.wav", y, sr)
sf.write("denoised_2.wav", y_denoised, sr)

print("Готово! Сохранены файлы: original.wav и denoised.wav")

# === Визуализация ===
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr, alpha=0.5, label="Оригинал")
librosa.display.waveshow(y_denoised, sr=sr, color='r', alpha=0.5, label="После шумоподавления")
plt.title("Сравнение сигналов")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write as write_wav

# Параметры
TONE_FREQ = 800  # Гц
SAMPLE_RATE = 22050  # 176128 точек = 8 секунд

DOT_DURATION = int(0.04 * SAMPLE_RATE)              # . (0.01 с)
DASH_DURATION = int(SAMPLE_RATE * 0.15)         # - (0.03 с)
INTRA_CHAR_SPACE = int(SAMPLE_RATE * 0.055)         # между точками и тире
INTER_CHAR_SPACE = int(SAMPLE_RATE * 0.095)    # между буквами
WORD_SPACE = int(SAMPLE_RATE * 0.35)           # между словами (0.33 с)

# Генерация синусоиды
def generate_tone(duration, freq=TONE_FREQ, sample_rate=SAMPLE_RATE):
    t = np.linspace(0, duration / sample_rate, duration, endpoint=False)
    tone = np.sin(2 * np.pi * freq * t)
    return tone.astype(np.float32)

# Преобразование азбуки Морзе в аудиосигнал (numpy)
def morse_to_audio_np(morse_code: str) -> np.ndarray:
    signal = np.array([], dtype=np.float32)

    for symbol in morse_code:
        if symbol == '.':
            tone = generate_tone(DOT_DURATION)
            pause = np.zeros(INTRA_CHAR_SPACE, dtype=np.float32)
            signal = np.concatenate((signal, tone, pause))
        elif symbol == '-':
            tone = generate_tone(DASH_DURATION)
            pause = np.zeros(INTRA_CHAR_SPACE, dtype=np.float32)
            signal = np.concatenate((signal, tone, pause))
        elif symbol == ' ':
            signal = np.concatenate((signal, np.zeros(INTER_CHAR_SPACE, dtype=np.float32)))
        elif symbol == '|':
            signal = np.concatenate((signal, np.zeros(WORD_SPACE, dtype=np.float32)))
        else:
            continue  # игнорируем посторонние символы

    return signal

# Дополнение до ровно 8 секунд
def pad_to_176128(signal):
    signal = np.asarray(signal, dtype=np.float32)
    if len(signal) < 176128:
        padding = np.zeros(176128 - len(signal), dtype=np.float32)
        signal = np.concatenate([signal, padding])
    else:
        signal = signal[:176128]
    return signal

# Визуализация
def plot_signal(signal, sample_rate=SAMPLE_RATE):
    time = np.linspace(0, len(signal)/sample_rate, len(signal))
    plt.figure(figsize=(15, 3))
    plt.plot(time, signal)
    plt.title("Morse Code Signal")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Сохранение в WAV
def save_as_wav(filename: str, signal: np.ndarray, sample_rate=SAMPLE_RATE):
    signal_int16 = np.int16(signal * 32767)  # Преобразование float32 → int16
    write_wav(filename, sample_rate, signal_int16)
```

```python
morse = " ..... ---|-.-- ----. ----. -.. ..-.. "  # SOS
signal = morse_to_audio_np(morse)
final_signal = pad_to_176128(signal)
plot_signal(final_signal)
save_as_wav("gen_signal.wav", final_signal)
print("Сигнал сохранён в sos_morse.wav")
```

```python
import librosa
import numpy as np
import pywt
import matplotlib.pyplot as plt
import soundfile as sf
import scipy.io.wavfile as wavfile
import librosa.display
from scipy.signal import butter, filtfilt, spectrogram
```

```python

def bandpass_filter(signal, fs, lowcut, highcut, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    filtered = filtfilt(b, a, signal)
    return filtered


def plot_spectrogram(signal, fs, title="Spectrogram"):
    plt.figure(figsize=(12, 4))
    D = librosa.amplitude_to_db(np.abs(librosa.stft(signal)), ref=np.max)
    librosa.display.specshow(D, sr=fs, x_axis='time', y_axis='hz', cmap='magma')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.tight_layout()
    plt.show()


def load_audio_librosa(filepath, sr=None):
    signal, sample_rate = librosa.load(filepath, sr=sr, mono=True)
    return signal, sample_rate


def wavelet_denoise(signal, wavelet='coif3', level=4):
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    uthresh = sigma * np.sqrt(2 * np.log(len(signal)))
    coeffs_thresh = [pywt.threshold(c, value=uthresh, mode='soft') for c in coeffs]
    return pywt.waverec(coeffs_thresh, wavelet)


def plot_signals(original, denoised, fs):
    time = np.linspace(0, len(original)/fs, num=len(original))
    plt.figure(figsize=(14, 5))
    plt.plot(time, original, label="Original", alpha=0.5)
    plt.plot(time, denoised, label="Denoised", alpha=0.7)
    plt.legend()
    plt.title("Сравнение исходного и очищенного сигнала")
    plt.xlabel("Время (сек)")
    plt.ylabel("Амплитуда")
    plt.tight_layout()
    plt.show()


def save_to_wav(filename, signal, fs):
    # Ограничим амплитуду перед сохранением
    signal = signal / np.max(np.abs(signal))
    sf.write(filename, signal, fs)
    print(f"Файл сохранён: {filename}")


# === ЗАПУСК ===
if __name__ == "__main__":
    input_file = "denoised_2.wav"
    signal, fs = load_audio_librosa(input_file)
    

    
    # Строим спектрограмму
    frequencies, times, Sxx = spectrogram(signal, fs)
    
    # Находим диапазон частот с максимальной амплитудой
    max_amplitude_freq_range = np.argmax(np.max(Sxx, axis=1))  # Частота с наибольшей амплитудой
    max_amplitude = np.max(Sxx)
    
    # Получаем диапазон частот для максимальной амплитуды
    # Например, будем смотреть на частоты в пределах 10% от максимальной амплитуды
    threshold = 0.03 * max_amplitude
    freq_range = frequencies[np.max(Sxx, axis=1) > threshold]

    denoised = wavelet_denoise(signal)

    
    # Обрезаем до совпадения длины
    min_len = min(len(signal), len(denoised))
    signal = signal[:min_len]
    denoised = denoised[:min_len]
    
    filtered = bandpass_filter(denoised, fs, lowcut=freq_range[0], highcut=freq_range[-1])

    # Визуализация временных сигналов
   # plot_signals(signal, denoised, fs)

    # Визуализация спектрограмм
    plot_spectrogram(filtered, fs, title="итоговая спектрограмма")
   # plot_spectrogram(denoised, fs, title="Спектрограмма после вейвлет-фильтрации")

    # Сохраняем результат
    # save_to_wav("denoised_output.wav", denoised, fs)
    # save_to_wav("filtered_output.wav", filtered, fs)
```

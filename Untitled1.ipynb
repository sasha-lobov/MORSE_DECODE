## предобработка

```python

def norm_signal(y, plot=True):

    min_val = np.min(y)
    max_val = np.max(y)
    
    if max_val == min_val:
        return np.zeros_like(y)  # если сигнал плоский — просто нули

    y_norm = 2 * (y - min_val) / (max_val - min_val) - 1
    x_norm = np.linspace(0, 8, num=len(y))
    
    if plot is True:
        plt.figure(figsize=(10, 4))
        plt.plot(x_norm, y_norm, color='red')
        plt.title("после нормализации")
        plt.show()
        
    return y_norm


def interpol(y, sr, N=2000, plot=True): # интерполяция

    x = np.linspace(0, len(y) / sr, num=len(y))  # Временная шкала
    
    interp_func = interp1d(x, y, kind='linear')  # можно заменить на 'cubic' для сглаживания
    x_resampled = np.linspace(x.min(), x.max(), N)
    y_resampled = interp_func(x_resampled)

    if plot is True:
        plt.figure(figsize=(10, 4))
        plt.plot(x_resampled, y_resampled, color='red')
        plt.title("после интерполяции")
        plt.show()
        
    return y_resampled


def noise_sub(y, sr, noise_duration=0.2, n_fft=2048, hop_length=512, save=False, plot=True): # вычиталка шума
    
    # преобразование STFT
    S_full = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
    S_mag, S_phase = np.abs(S_full), np.angle(S_full)
    
    # оценка спектра шума
    noise_frames = int(noise_duration * sr / hop_length)
    noise_profile = np.mean(S_mag[:, :noise_frames], axis=1, keepdims=True)
    
    # спектральное вычитание
    S_denoised_mag = np.maximum(S_mag - noise_profile, 0)
    S_denoised = S_denoised_mag * np.exp(1j * S_phase)
    
    # обратное преобразование
    y_denoised = librosa.istft(S_denoised, hop_length=hop_length)
    
    # сохранение аудио
    if save is True:
        #sf.write("original.wav", y, sr)
        sf.write("denoised_2.wav", y_denoised, sr)
        print("Готово! Сохранены файлы: original.wav и denoised.wav")
        
    if plot is True:
        # визуализация
        plt.figure(figsize=(10, 4))
        librosa.display.waveshow(y, sr=sr, alpha=0.5, label="Оригинал")
        librosa.display.waveshow(y_denoised, sr=sr, color='r', alpha=0.5, label="После шумоподавления")
        plt.title("Сравнение сигналов")
        plt.legend()
        plt.show()
        
    return y_denoised
    

def wavelet_denoise(y, wavelet='coif3', level=4):
    
    coeffs = pywt.wavedec(y, wavelet, level=level)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    uthresh = sigma * np.sqrt(2 * np.log(len(y)))
    coeffs_thresh = [pywt.threshold(c, value=uthresh, mode='soft') for c in coeffs]
    return pywt.waverec(coeffs_thresh, wavelet)


def bandpass_filter(y, sr, lowcut, highcut, order=4):
    nyquist = 0.5 * sr

    # Защита от перепутанных или слишком близких значений
    lowcut, highcut = sorted([lowcut, highcut])
    if abs(highcut - lowcut) < 10:  # менее 10 Гц — это странно
        print(f"⚠️ Частоты слишком близки: {lowcut}–{highcut}, расширяем")
        delta = 10
        lowcut = max(1, lowcut - delta)
        highcut = min(nyquist - 1, highcut + delta)

    low = lowcut / nyquist
    high = highcut / nyquist

    # Безопасность
    high = min(high, 0.999)
    if low >= high:
        raise ValueError(f"❌ Ошибка: low ({low}) >= high ({high})")

    b, a = butter(order, [low, high], btype='band')
    filtered = filtfilt(b, a, y)
    return filtered


def spectr_filt(y, sr, range_hz=0.07, plot=True):
    # получаю хар-ки спектрограммы
    frequencies, times, Sxx = spectrogram(y, sr)
    
    # диапазон частот с максимальной амплитудой
    max_amplitude_freq_range = np.argmax(np.max(Sxx, axis=1))  # Частота с наибольшей амплитудой
    max_amplitude = np.max(Sxx)
    
    # диапазон частот для максимальной амплитуды
    threshold = range_hz * max_amplitude
    freq_range = frequencies[np.max(Sxx, axis=1) > threshold]
    
    if len(freq_range) < 2 or freq_range[0] >= freq_range[-1]:
        # freq_range[0] = freq_range[0] - freq_range[0]*0.02
        # freq_range[-1] = freq_range[-1] + freq_range[-1]*0.02
        freq_range = [300, 3000]
        
    denoised = wavelet_denoise(y)

    # обрезаю до совпадения длины
    min_len = min(len(y), len(denoised))
    y = y[:min_len]
    denoised = denoised[:min_len]
    
    filtered = bandpass_filter(denoised, sr, lowcut=freq_range[0], highcut=freq_range[-1])

    if plot is True:
        plt.figure(figsize=(10, 4))
        D = librosa.amplitude_to_db(np.abs(librosa.stft(filtered)), ref=np.max)
        librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='hz', cmap='magma')
        plt.colorbar(format='%+2.0f dB')
        plt.title("Спектрограмма")
        plt.tight_layout()
        plt.show()
        
    return filtered


def env_filt(y, sr, gaus_filt=False, save=False, plot=True):
    
    # расчёт энвелопа через скользящее RMS
    frame_duration = 0.05  # окно 50 мс
    frame_size = int(sr * frame_duration)
    envelope = np.sqrt(np.convolve(y ** 2, np.ones(frame_size) / frame_size, mode='same'))

    if gaus_filt is True:
        sigma = 300  # Чем больше, тем сильнее сглаживание
        envelope = gaussian_filter1d(envelope, sigma=sigma)
    
    if plot is True:
        t = np.linspace(0, len(y) / sr, len(y))
    
        plt.figure(figsize=(15, 8))
        
        plt.subplot(2, 1, 1)
        plt.plot(t, y, label='Original Signal')
        plt.title('Оригинальный сигнал')
        
        plt.subplot(2, 1, 2)
        plt.plot(t, envelope, label='Envelope', color='orange')
        plt.title('Энвелоп')
        plt.legend()

    if save is True:
        output_file = 'env_signal.wav'
        sf.write(output_file, cleaned_signal, sr)
        print(f"Сигнал сохранён в файл: {output_file}")
    
    return envelope


def save_audio(audio_array, path):
    np.savetxt(path, audio_array, fmt="%.6f")


def load_audio(path):
    return np.loadtxt(path, dtype=np.float32)
```

## генерация сигнала

```python

# Параметры
TONE_FREQ = 1000  # Гц
SAMPLE_RATE = 22050  # 176128 точек = 8 секунд

DOT_DURATION = int(0.04 * SAMPLE_RATE)              # . (0.01 с)
DASH_DURATION = int(SAMPLE_RATE * 0.15)         # - (0.03 с)
INTRA_CHAR_SPACE = int(SAMPLE_RATE * 0.055)         # между точками и тире
INTER_CHAR_SPACE = int(SAMPLE_RATE * 0.095)    # между буквами
WORD_SPACE = int(SAMPLE_RATE * 0.35)           # между словами (0.33 с)

# Генерация синусоиды
def generate_tone(duration, freq=TONE_FREQ, sample_rate=SAMPLE_RATE):
    t = np.linspace(0, duration / sample_rate, duration, endpoint=False)
    tone = np.sin(2 * np.pi * freq * t)
    return tone.astype(np.float32)

# Преобразование азбуки Морзе в аудиосигнал (numpy)
def morse_to_audio_np(morse_code: str) -> np.ndarray:
    signal = np.array([], dtype=np.float32)

    for symbol in morse_code:
        if symbol == '.':
            tone = generate_tone(DOT_DURATION)
            pause = np.zeros(INTRA_CHAR_SPACE, dtype=np.float32)
            signal = np.concatenate((signal, tone, pause))
        elif symbol == '-':
            tone = generate_tone(DASH_DURATION)
            pause = np.zeros(INTRA_CHAR_SPACE, dtype=np.float32)
            signal = np.concatenate((signal, tone, pause))
        elif symbol == ' ':
            signal = np.concatenate((signal, np.zeros(INTER_CHAR_SPACE, dtype=np.float32)))
        elif symbol == '|':
            signal = np.concatenate((signal, np.zeros(WORD_SPACE, dtype=np.float32)))
        else:
            continue  # игнорируем посторонние символы

    return signal

# Дополнение до ровно 8 секунд
def pad_to_176128(signal):
    signal = np.asarray(signal, dtype=np.float32)
    if len(signal) < 176128:
        padding = np.zeros(176128 - len(signal), dtype=np.float32)
        signal = np.concatenate([signal, padding])
    else:
        signal = signal[:176128]
    return signal

# Визуализация
def plot_signal(signal, sample_rate=SAMPLE_RATE):
    time = np.linspace(0, len(signal)/sample_rate, len(signal))
    plt.figure(figsize=(15, 3))
    plt.plot(time, signal)
    plt.title("Morse Code Signal")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Сохранение в WAV
def save_as_wav(filename: str, signal: np.ndarray, sample_rate=SAMPLE_RATE):
    signal_int16 = np.int16(signal * 32767)  # Преобразование float32 → int16
    write_wav(filename, sample_rate, signal_int16)
```

```python
# конвертер: код Морзе -> последовательность фрагментов
def convert_seq(code, a1,a2,a3,a4,a5,a6,a7,a8): 
    
    seq = []
    for i in range(0, len(code)):
        
        segm = {"waveform": 1, "class": 1, "length": 1}
        
        if i == 0:
            if code[i] == '.' and (code[i+1] == '.' or code[i+1] == '-'):
                segm["waveform"] = a2
                segm["class"] = "dot_2"
                segm["length"] = len(a2)
            elif code[i] == '.' and (code[i+1] == ' ' or code[i+1] == '|'):
                segm["waveform"] = a1
                segm["class"] = "dot_1"
                segm["length"] = len(a1) 
            elif code[i] == '-' and (code[i+1] == '.' or code[i+1] == '-'):
                segm["waveform"] = a6
                segm["class"] = "dash_2"
                segm["length"] = len(a6) 
            elif code[i] == '-' and (code[i+1] == ' ' or code[i+1] == '|'):
                segm["waveform"] = a5
                segm["class"] = "dash_1"
                segm["length"] = len(a5)                 
                
        elif i > 0 and i < len(code)-1:
            if (code[i-1] == '.' or code[i-1] == '-') and code[i] == '.' and (code[i+1] == '.' or code[i+1] == '-'): # 
                segm["waveform"] = a4
                segm["class"] = "dot_4"
                segm["length"] = len(a4) 
            if (code[i-1] == ' ' or code[i-1] == '|') and code[i] == '.' and (code[i+1] == '.' or code[i+1] == '-'): #
                segm["waveform"] = a2
                segm["class"] = "dot_2"
                segm["length"] = len(a2)
            if (code[i-1] == ' ' or code[i-1] == '|') and code[i] == '.' and (code[i+1] == ' ' or code[i+1] == '|'): #
                segm["waveform"] = a1
                segm["class"] = "dot_1"
                segm["length"] = len(a1)
            if (code[i-1] == '.' or code[i-1] == '-') and code[i] == '.' and (code[i+1] == ' ' or code[i+1] == '|'): #
                segm["waveform"] = a3
                segm["class"] = "dot_3"
                segm["length"] = len(a3)
            if (code[i-1] == '.' or code[i-1] == '-') and code[i] == '-' and (code[i+1] == '.' or code[i+1] == '-'): #
                segm["waveform"] = a8
                segm["class"] = "dash_4"
                segm["length"] = len(a8) 
            if (code[i-1] == ' ' or code[i-1] == '|') and code[i] == '-' and (code[i+1] == '.' or code[i+1] == '-'): #
                segm["waveform"] = a6
                segm["class"] = "dash_2"
                segm["length"] = len(a6)
            if (code[i-1] == ' ' or code[i-1] == '|') and code[i] == '-' and (code[i+1] == ' ' or code[i+1] == '|'): 
                segm["waveform"] = a5
                segm["class"] = "dash_1"
                segm["length"] = len(a5)
            if (code[i-1] == '.' or code[i-1] == '-') and code[i] == '-' and (code[i+1] == ' ' or code[i+1] == '|'): 
                segm["waveform"] = a7
                segm["class"] = "dash_3"
                segm["length"] = len(a7)

        elif i == len(code)-1:
            if code[i] == '.' and (code[i-1] == '.' or code[i-1] == '-'):
                segm["waveform"] = a3
                segm["class"] = "dot_3"
                segm["length"] = len(a3)
            elif code[i] == '.' and (code[i-1] == ' ' or code[i-1] == '|'):
                segm["waveform"] = a1
                segm["class"] = "dot_1"
                segm["length"] = len(a1) 
            elif code[i] == '-' and (code[i-1] == '.' or code[i-1] == '-'):
                segm["waveform"] = a7
                segm["class"] = "dash_3"
                segm["length"] = len(a7) 
            elif code[i] == '-' and (code[i-1] == ' ' or code[i-1] == '|'):
                segm["waveform"] = a5
                segm["class"] = "dash_1"
                segm["length"] = len(a5) 

        if segm['length'] > 1:
            seq.append(segm)  
        segm = {"waveform": 1, "class": 1, "length": 1}

    return seq

    
```

```python
# сохранение/загрузка джсонов
import json

def save_label(label, path):
    """
    Сохраняет список фрагментов в .json, где waveform сериализуется как обычный список.
    """
    serializable = [
        {
            "waveform": frag["waveform"].tolist(),  # из numpy в список
            "class": frag["class"],
            "length": frag["length"]
        }
        for frag in label
    ]
    with open(path, "w") as f:
        json.dump(serializable, f, indent=2)


def load_label(path):
    with open(path, "r") as f:
        raw = json.load(f)
    return [
        {
            "waveform": np.array(frag["waveform"], dtype=np.float32),
            "class": frag["class"],
            "length": frag["length"]
        }
        for frag in raw
    ]
```
